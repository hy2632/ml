```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Path of the file to read
museum_filepath = "input/museum_visitors.csv"

# Fill in the line below to read the file into a variable museum_data
museum_data = pd.read_csv(museum_filepath, index_col="Date", parse_dates=True)
```

```python
museum_data = museum_data[['Avila Adobe', 'Firehouse Museum', 'Chinese American Museum',
       'America Tropical Interpretive Center']]
```

```python
# Fill in the line below: How many visitors did the Chinese American Museum 
# receive in July 2018?
ca_museum_jul18 = int(museum_data['2018-07-01']['Chinese American Museum'])

# Fill in the line below: In October 2018, how many more visitors did Avila 
# Adobe receive than the Firehouse Museum?
avila_oct18 = int(museum_data['2018-10-01']['Avila Adobe'] - museum_data['2018-10-01']['Firehouse Museum'])
```

```python
ca_museum_jul18
```

```python
avila_oct18
```

```python
# Line chart showing the number of visitors to each museum over time
plt.figure(figsize=(12,6))
sns.lineplot(data = museum_data)
plt.title("Monthly Visitors to Los Angeles City Museums")
```
